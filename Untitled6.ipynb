```python
# ============================================================
# Fake Job Posting Detection Using Machine Learning
# ============================================================

# Import Libraries
import pandas as pd
import numpy as np
import re
import string

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel("fake_job_postings.csv.xlsx")

print("Dataset Shape:", df.shape)
print(df.head())

# ============================================================
# Handle Missing Values
# ============================================================

df.fillna('', inplace=True)

# ============================================================
# Combine Important Text Columns
# ============================================================

text_columns = [
    'title',
    'company_profile',
    'description',
    'requirements',
    'benefits'
]

df['text'] = df[text_columns].agg(' '.join, axis=1)

# ============================================================
# Text Cleaning Function
# ============================================================

def clean_text(text):

    text = text.lower()

    text = re.sub(r"http\S+", "", text)

    text = re.sub(r"\d+", "", text)

    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    text = re.sub(r"\s+", " ", text)

    return text.strip()

df['text'] = df['text'].apply(clean_text)

# ============================================================
# Feature & Target
# ============================================================

X = df['text']

y = df['fraudulent']

# ============================================================
# TF-IDF Feature Extraction
# ============================================================

vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=10000
)

X = vectorizer.fit_transform(X)

# ============================================================
# Train Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# ============================================================
# Function to Evaluate Models
# ============================================================

def evaluate_model(name, model):

    model.fit(X_train, y_train)

    prediction = model.predict(X_test)

    print("="*60)
    print(name)
    print("="*60)

    print("Accuracy :", accuracy_score(y_test, prediction))
    print("Precision:", precision_score(y_test, prediction))
    print("Recall   :", recall_score(y_test, prediction))
    print("F1 Score :", f1_score(y_test, prediction))

    print("\nConfusion Matrix")
    print(confusion_matrix(y_test, prediction))

    print("\nClassification Report")
    print(classification_report(y_test, prediction))

# ============================================================
# Logistic Regression
# ============================================================

lr = LogisticRegression(max_iter=1000)

evaluate_model(
    "Logistic Regression",
    lr
)

# ============================================================
# Naive Bayes
# ============================================================

nb = MultinomialNB()

evaluate_model(
    "Naive Bayes",
    nb
)

# ============================================================
# Random Forest
# ============================================================

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

evaluate_model(
    "Random Forest",
    rf
)

# ============================================================
# Support Vector Machine
# ============================================================

svm = LinearSVC()

evaluate_model(
    "Support Vector Machine",
    svm
)

# ============================================================
# Train Final Model
# ============================================================

final_model = LogisticRegression(max_iter=1000)

final_model.fit(X_train, y_train)

# ============================================================
# Prediction Function
# ============================================================

def predict_job(title,
                company_profile,
                description,
                requirements,
                benefits):

    text = (
        title + " " +
        company_profile + " " +
        description + " " +
        requirements + " " +
        benefits
    )

    text = clean_text(text)

    text = vectorizer.transform([text])

    prediction = final_model.predict(text)[0]

    probability = None

    if hasattr(final_model, "predict_proba"):
        probability = final_model.predict_proba(text)[0]

    if prediction == 1:
        print("\nPrediction : Fraudulent Job Posting")
    else:
        print("\nPrediction : Legitimate Job Posting")

    if probability is not None:
        print("Probability:", probability)

# ============================================================
# Sample Prediction
# ============================================================

predict_job(

title="Data Scientist",

company_profile="""
ABC Technologies is a leading AI company.
""",

description="""
Looking for a Data Scientist with experience in
Python, Machine Learning, SQL, NLP,
Deep Learning and Statistics.
""",

requirements="""
Bachelor's Degree
2+ years experience
Python
SQL
Machine Learning
""",

benefits="""
Health Insurance
Flexible Working Hours
Paid Leave
"""
)
```
